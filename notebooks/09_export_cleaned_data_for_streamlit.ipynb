# Notebook 09: Export Cleaned Data for Streamlit

## Objective
This notebook prepares and exports final datasets for Streamlit dashboard development.

### Key Tasks:
- Organize datasets by dashboard module
- Remove unnecessary columns
- Ensure consistency in naming and structure
- Export optimized CSV files

These datasets will be directly used in the Streamlit dashboard.

```python
import pandas as pd

pd.set_option("display.max_columns", None)
```

## Load all Data

```python
# Load datasets
ncd = pd.read_csv("../data/featured_nassau_data.csv")

route_state = pd.read_csv("../data/route_state_summary.csv")
route_region = pd.read_csv("../data/route_region_summary.csv")
route_benchmark = pd.read_csv("../data/route_efficiency_benchmark.csv")
state_bottleneck = pd.read_csv("../data/state_bottleneck_analysis.csv")
ship_mode = pd.read_csv("../data/ship_mode_detailed_analysis.csv")

print("All datasets loaded successfully.")
```

## CLEAN MAIN DATASET

### Prepare Main Dataset
Keep only necessary columns for dashboard performance.

```python
# Select required columns
dashboard_main = ncd[[
    "Order ID",
    "Order Date",
    "Ship Date",
    "Ship Mode",
    "Factory",
    "State/Province",
    "Region",
    "Route_State",
    "Route_Region",
    "Shipping Lead Time",
    "Delay Status",
    "Shipping Speed Category",
    "Sales",
    "Units",
    "Cost",
    "Gross Profit"
]].copy()
```

```python
dashboard_main.head()
```

## Ensure DateTime

```python
dashboard_main["Order Date"] = pd.to_datetime(dashboard_main["Order Date"])
dashboard_main["Ship Date"] = pd.to_datetime(dashboard_main["Ship Date"])
```

## Sort Data

```python
dashboard_main = dashboard_main.sort_values(by="Order Date")
```

## Final Datasets

### Final Dataset Preparation

```python
# Route leaderboard dataset
route_leaderboard = route_benchmark.copy()

# State bottleneck dataset
state_map = state_bottleneck.copy()

# Ship mode dataset
ship_mode_final = ship_mode.copy()
```

## SAVE ALL FILES

```python
dashboard_main.to_csv("../data/processed/dashboard_main.csv", index=False)
route_leaderboard.to_csv("../data/processed/dashboard_route_leaderboard.csv", index=False)
route_region.to_csv("../data/processed/dashboard_region_summary.csv", index=False)
ship_mode_final.to_csv("../data/processed/dashboard_ship_mode.csv", index=False)
state_map.to_csv("../data/processed/dashboard_state_map.csv", index=False)

print("All dashboard datasets saved successfully.")
```

## Final Check

```python
print("Dashboard Main Shape:", dashboard_main.shape)
print("Route Leaderboard Shape:", route_leaderboard.shape)
print("Region Summary Shape:", route_region.shape)
print("Ship Mode Shape:", ship_mode_final.shape)
print("State Map Shape:", state_map.shape)
```
