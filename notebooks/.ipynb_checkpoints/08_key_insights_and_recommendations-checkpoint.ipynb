## Key Insights

### 1. Route Efficiency is not uniform across the network
The route benchmarking results show that shipping performance varies meaningfully across factory-to-state routes. Some routes maintain lower average lead times and better consistency, while others show relatively weaker efficiency.

### 2. High-volume routes deserve special operational attention
Routes or states with both high shipment volume and weaker performance represent the biggest operational risk. These locations can create broader customer experience issues because even small inefficiencies affect a larger number of orders.

### 3. Geographic bottlenecks are important for logistics planning
The bottleneck analysis highlights that some states and regions face higher average lead times than others. These areas should be prioritized for route planning review, capacity balancing, and process improvement.

### 4. Shipping mode has a clear relationship with delivery speed
Ship mode analysis confirms that faster shipping categories generally produce lower lead times, while slower modes tend to have longer delivery durations. This supports the expected service-speed structure of the logistics system.

### 5. Delay frequency should be monitored as an operational KPI
Even when average lead time appears acceptable, high delay rates in certain shipping modes or geographies can reduce reliability. Delay Frequency should therefore be treated as an important operational monitoring KPI.

### 6. Efficiency should be evaluated using both speed and consistency
A route with low average lead time but high variability may still create service inconsistency. For this reason, route efficiency should be assessed using both average lead time and lead time variability.

## Business Recommendations

### 1. Prioritize bottom-performing routes for operational review
The least efficient routes should be investigated first to identify the causes of slower delivery performance. Possible causes may include factory-to-destination distance, ship mode allocation, route congestion, or planning inefficiencies.

### 2. Focus improvement efforts on high-volume bottleneck states
States with both high shipment volume and poor performance should receive the highest priority because improvements there can create the largest business impact.

### 3. Use ship mode strategy more deliberately
If slower modes are driving delay frequency in key regions, the organization should consider whether certain customer segments, locations, or product categories require different shipping policies.

### 4. Monitor route variability, not just average lead time
Operational teams should track lead time consistency along with average delivery speed. Routes with unstable performance may require process controls even if their average lead time seems acceptable.

### 5. Build a logistics monitoring dashboard for ongoing visibility
A Streamlit dashboard can help decision-makers track route efficiency, state-level bottlenecks, ship mode performance, and delay metrics in a centralized and user-friendly way.

### 6. Use route-level intelligence to support scalability
As shipment volume grows, route-level visibility will become even more important. Early identification of weak routes can help the organization scale logistics operations more efficiently.

## Executive Summary

This project transformed raw shipment records into route-level logistics intelligence for Nassau Candy Distributor. The analysis identified meaningful differences in shipping efficiency across routes, states, regions, and ship modes.

The findings suggest that logistics performance should not be evaluated only at an overall level. Instead, route-level and geographic-level visibility is necessary to identify weak routes, bottleneck areas, and reliability risks.

The project also shows the value of using operational KPIs such as:
- Shipping Lead Time
- Average Lead Time
- Route Volume
- Delay Frequency
- Route Efficiency Score

Overall, this analysis provides a strong foundation for a data-driven logistics dashboard that can support better planning, improved delivery reliability, and smarter operational decision-making.
